```python
import sqlite3

conn = sqlite3.connect('data/movies.sqlite')
db = conn.cursor()
```

```python
query = "SELECT COUNT(*) from directors d"
db.execute(query)
result = db.fetchone()
print(result[0])
```

```python
query = "select d.name  from directors d order by d.name"
db.execute(query)
results = db.fetchall()
#print(results)

print([each_director[0] for each_director in results])
```

```python
query = "select m.title  from movies m WHERE UPPER(m.title) LIKE '% LOVE %'\
OR    UPPER(m.title) LIKE 'LOVE %'\
OR    UPPER(m.title) LIKE '% LOVE'\
OR    UPPER(m.title) = 'LOVE'"
db.execute(query)
results = db.fetchall()
#print(results)

print([each_movie[0] for each_movie in results])
```
